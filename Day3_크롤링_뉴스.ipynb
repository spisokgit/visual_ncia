```python
from bs4 import BeautifulSoup
import requests
```

```python
r=requests.get("https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001")
r
```

```python
c = r.content
soup=BeautifulSoup(c, "html.parser")
soup
```

```python
# 크롬 F12(c f i)에서  elements에서 화살표 눌려서 해당 뉴스 소스코드 특정,
# ctrl f해서  해당 id 중복여부 확인
# 기사내용(ul(class="type06_headline">li>dl>dd>span(class="lede")

all = soup.find("ul",{"class" : "type06_headline"}) # class 유일한지 여부 확인
print(f'type(all):{type(all)}')  #type(all)<class 'bs4.element.Tag'> 
all
```

```python
lst = []
dl=all.find_all("dl")
for item in dl:
    content = item.find("dd")
    lst.append(content.find("span", {"class":"lede"}).text)            
lst
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rc("font",family="Malgun Gothic")
```

```python
# 엑셀 파일 읽기

df = pd.read_excel('costco.xlsx', encoding='utf-8') # Shift+Tap!
```

```python
df_new = df[['성명', '판정일', '신장', '체중', '허리둘레']] # 에러를 고쳐보세요!
```

```python
# Type 을 index 로 하는 새로운 df 만들기 (엑셀의 pivot table 과 유사)

pivot_df = pd.pivot_table(df_new, index=['성명','판정일'], aggfunc=np.mean) # Shift+Tab!
pivot_df
```

```python
#이름 리턴
df["성명"]
```

```python
#이름의 중복을 제거하고 리턴
names=df["성명"].unique()
names
```

```python
df02=pivot_df.copy()
df02
```

```python
#강규응의 데이터 조회
data = df02.loc["강규응"]
data
```

```python
#강규응의 신장 조회
data["신장"]
```

```python
def name_plot(name):
    data = df02.loc[name]
    plt.plot(data["신장"])
```

```python
name_plot("강규응")
```

```python
#모든 이름의 개수
total_count=len(names)
total_count
```

```python
#모든 이름을 5로 나눈 몫 조회
total_row=total_count//5
total_row
```

```python
#total_count 가 5로 나눈 나머지가 0이 아니면
if total_count%5!=0:
    #total_row를 1증가
    total_row+=1
total_row
```

```python
#전체 그래프의 가로 세로
fig=plt.figure(figsize = (30, 30))
#total_count만큼 반복
for i in range(total_count):
    #그래프를 total_row줄 5칸으로 분리하고 i+1번째 그래프를 리턴함
    subplot=fig.add_subplot(total_row,5,i+1)
    #i번째 이름을 name에 저장
    name=names[i]
    #name의 정보를 리턴
    data = df02.loc[name]
    #신장 정보를 그래프를 그림
    subplot.plot(data["신장"])
    #이름으로 그래프 제목 설정
    subplot.set_title(name)

#전체 그래프 그리기
plt.show()
```

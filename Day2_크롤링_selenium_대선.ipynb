```python
# 보기좋게 bs4 처리
#from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
table = soup.find(id="table01")
print(f'type(table):{type(table)}')  #type(table):<class 'bs4.element.Tag'> 
table
```

```python
# 문자열로 변경하여 판다스로 처리하기
import pandas as pd
df=pd.read_html(str(table)) # 문자열로 바뀐 상태에서 문자열만 추출해 내어 데이터프레임형태로 list에 저장
print(f'type(df) : {type(df)}')
print(f'type(df[0]):{type(df[0])}')  # list안에 데이터프레임들이 있다.
df
```

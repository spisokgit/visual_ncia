
## DataFrame, Series

```python
# dict ==> pd
raw_data = {"first_name" : ["길동", "몽룡"],
            "last_name" : ["홍", "이"],
            "age" : [32,42],
            "city":["서울", "부산"]
           }                         # list, ndarray, dict ===> pd
```

```python
pd.DataFrame(raw_data).head()
```

```python
pd.DataFrame(raw_data, columns=["age", "city"]) #대소문자 구분하므로 Columns하면 안됨
```

```python
# column 순서바꾸고, 결즉치(NaN) 데이터 column 추가
df=pd.DataFrame(raw_data, columns=["last_name", "first_name", "age", "city", "deby"]) 
df
```

```python
print(df.first_name) # 속성접근 : column 1개일경우만 가능  column 2개이상보고싶을때 안됨
print(df["first_name"]) # 
df[["city", "first_name"]] # 2개 column보고싶을때 List안에 넘어야 한다.
```

```python
# 줄(raw)이름 : 보이는index 변경   # pd 슬라이싱에서 사용하는 iloc(안보이는 index), loc(줄이름 보이는index)
df.index = ["A","B"]
df
```

```python
# 줄이름, 줄순서로 줄 찾기
print(f'줄이름(보이는 index) 줄찾기\n {df.loc["B"]}\n')
print(f'줄순서(안보이는 index) 줄찾기\n {df.iloc[1]}') # 줄순서(iloc)는 0, 1, ,,,
```

```python
df1=df[df.city=="서울"] # boolean 색인 => groupby 
df1
```

```python
df[df.city=="서울"]['age'] # index와 
```

```python
df.age > 40
```

```python
# column명은 신규 추가도 되고, 있으면 수정도 됨. # 속성(.)으로 접근시 수정만 되고, 생성은 되지 않음
df["debt"]=df.age>40
df
```

```python
df[df.age>40] # 속성으로 접근
df[df["age"]>40] #동일함
```

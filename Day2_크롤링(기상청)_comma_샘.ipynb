```python
soup=BeautifulSoup(response.content,"html.parser")
soup
```

```python
table=soup.find("table",{"class":"table_develop3"})
table
```

```python
table.find_all("tr")
```

```python
for tr in table.find_all("tr"):
    print("="*50)
    print("tr:",tr)
    tds=tr.find_all("td")
    print("len(tds):",len(tds))
    print("tds:",tds)
    if len(tds)!=0:
        print("point:",tds[0].text)
        print("temperature:",tds[5].text)
        print("humidity:",tds[10].text)
    #for td in tds:
        #print("td:",td)
        #print("td.text:",td.text)
    print("="*50)
```

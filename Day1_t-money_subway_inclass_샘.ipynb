```python
new_data=[
    ['1월','1호선',"01","서울역",100,200,800,400],
    ["1월","1호선","02","용산역",300,700,200,500],
    ["1월","1호선","03","신림역",200,800,100,200]
]
```

```python
new_data
```

```python
for row in new_data:
    print("row:",row)
    for j in range(2):
        print("j=",j)
        print("j*2+4:", j*2+4)
        a=row[j*2+4]
        print("a:",a)
        print("0:mx[%s]:%s"%(j,mx[j]))
        if a>mx[j]:
            mx[j]=a
            mx_station[j]=row[3]
        print("1:mx[%s]:%s"%(j,mx[j]))
    print("="*50)
```

```python
data=[
    [1,2,3,100],
    [4,5,6,200]
]
```

```python
s_in=[0]*2
s_out=[0]*2
```

```python
for row in data:
    print("row:",row)
    for i in range(2):
        print("i=",i)
        print("i*2:",i*2)
        print("1+i*2:",1+i*2)
        a1=row[i*2]
        a2=row[1+i*2]
        print("a1:",a1,":a2:",a2)
        print("s_in[%s]:%s" %(i, s_in[i]))
        print("s_out[%s]:%s" %(i, s_out[i]))
        s_in[i]+=a1
        s_out[i]+=a2
        print("s_in[%s]:%s" %(i, s_in[i]))
        print("s_out[%s]:%s" %(i, s_out[i]))
        print("="*30)
```

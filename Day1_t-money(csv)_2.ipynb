## 모든 지하철역에서 시간대별 승하차 인원을 모두 더하면 

```python
import csv # 읽어만 보겠다. # pandas는 메모리에 계속 가지고 있으나, csv는 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

f = open("subwaytime.csv")
data  = csv.reader(f)
next(data) # 한줄recode  건너뛰기
next(data) # 또 한줄 recode 건너뛰기

# 시간대별 승하차 인원 저장할 리스트
s_in = [0]*24
s_out = [0]*24

for row in data: # data 문자
    #column 4번째부터 int형으로 변환하여 덮어쓰기
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4+i*2]    # 승차 
        s_out[i] += row[5+i*2]   # 하차
print(f's_in : \n{s_in}')
print(f's_out : \n{s_out}')
```

```python
#한글 
plt.rc('font', family='Malgun Gothic')

plt.figure(figsize=(10,4)) # dpi

#승차, 하차
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend(fontsize=12, loc='upper right') # legend position
#plt.xticks(list(map(lambda x : 4+x, range(24)))) # xticks : x축 # rotation 세로
plt.title('지하철 시간대별 승하차 인원 추이')

#plt.xticks(np.linspace(4, 27, 24)) # 4 처음 숫자 위치가 틀림 
plt.xticks(range(24), range(4,28)) # 갯수, 값
#plt.ylims()
plt.show()
```

```python
time=[]
for t in range(4,28):
    if t<25:
        time.append(t)
    else:
        time.append(t-24)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24),time) # 갯수, 값
plt.show()
```

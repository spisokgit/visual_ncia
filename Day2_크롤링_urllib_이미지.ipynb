```python
import urllib.request
```

```python
url="https://shop-phinf.pstatic.net/20190315_12/k2ci_1552615968460XOb7e_JPEG/40670071852290935_-1288493031.jpg"
```

```python
url.split("/")
```

```python
filename = url.split("/")[-1]
#filename  이름만 정함
```

```python
# binary 가져옴
mem = urllib.request.urlopen(url).read()  
mem # binary
```

```python
with open(filename, mode="wb") as file: # filename 변수
    file.write(mem) # 폴더 위치에 그림파일이 저장됨
```
